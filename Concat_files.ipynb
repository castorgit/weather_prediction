```python
# Concat files
import pandas as pd
```

```python

txt = './YHZ_hourly/climate-hourly ('
out = './YHZ_hourly/YHZ_hourly.csv'

file_name = txt +"0).csv"
file_out = pd.read_csv(file_name)
for i in range (1, 38):
    file_name = txt +str(i)+').csv'
    df = pd.read_csv(file_name)
    file_out = pd.concat([file_out, df], axis=0)
file_out.to_csv(out, index=False)
```
